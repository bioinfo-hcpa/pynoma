## Batch search

```python
import pynoma


searches = [pynoma.GeneSearch(2,"idua"), 
            pynoma.GeneSearch(2,"ace2"), 
            pynoma.GeneSearch(2,"mmp11"), 
            pynoma.GeneSearch(2,"id4")]

pynoma.batch_search(searches)
```
